```r
library(RGPR)
library(rChoiceDialogs)
```

```r
dt1_file <- "./Project12/Project12.dt1"

data <- readGPR(dt1_file)
n <- nrow(data@data)

print(data)
print(n)

plot(data)
```

```r
data_dewow <- dewow(data)

data_filtered <- filter(data_dewow, f = c(10, 100), type = "bandpass")

data_corrected <- t0Cor(data_filtered)

data_img <- kirchhoff(data_corrected)

vel_analysis <- velan(data_corrected, v = seq(0.05, 0.3, by = 0.01))

plot(data_img)
plot(vel_analysis)

writeGPR(data_img, file = "./Project12/Processed_Project12.dt1")
```
